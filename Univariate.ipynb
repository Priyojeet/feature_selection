```python
array = new_data.values
```

```python
X1 = array[:,0:2]
X2 = array[:, 3:14]
X = np.concatenate((X1, X2), axis=1)

Y = array[:,2]
```

```python
Y
```

```python
X
```

```python
test = SelectKBest(score_func=chi2, k=10)
```

```python
fit = test.fit(X, Y)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
features = fit.transform(X)
```

```python
print(features[0:6,:])
```

```python
for i in fit.scores_:
    print(i)
    j = np.float64(i).astype(str)
    print(i,"---",j)
```
